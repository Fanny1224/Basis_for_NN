# Weak Gravitational Lensing and Its Cosmological Applications  

#### **Def of WGL**  
the shearing(剪切) of distant galaxy images due to the differential deflection of neighboring light ray  
typical ellipticity: 1%(relatively very subtle)  

#### **Cosmic shear: lensing by large-scale structure in the universe**    
popularity: the signal is a direct measure of the projected matter power spectrum <font color=red>over a redshift range  
<font color=black>→ determine the cosmological parameters with high precision & powerful probes of dark energy and modified gravity theories  

### <font color=green>**How to measure shear**  
- **<font color=green>Mathematical descriptions**
    - the effect of lensing is a remap of thr source's surface brightness distribution.
    - The remap/distortion can be described by the <font color=red>distortion matrix  
    <img src="./fig/matrix.png" width="40%">

    - <font color=red>lensing convergence $\kappa(\theta)$ <font color=black>is determined by the lensing potential field $\psi$, which also determines the coefficients in the distortion matrix.  
    - convergence-maginification of the field
    - <font color=red>shear-ellipticity <font color=black>of the source  
      $e^{obs}=\frac{e^{\mathrm{int}}+\gamma}{1+\gamma e^{\mathrm{int}}}$, where $e^{obs}\approx\gamma$ in weak lensing, and the ellipticity is valued by $e\approx ((1-b/a)(1+b/a))e^{2i\alpha}$ ($\alpha$ is the orientation angle).

- **<font color=green>Weak lensing pipeline**
    - <font color=red>Obiect detection  
      <font color=black>on individual exposures or on a stacked image  
      tell galaxies from stars  
      then correct the effects induced by the camera and quantify the parameters of their shape  
    - <font color=red>Point Spread Function (PSF) estimation  
      <font color=black>In space-based data, the finite size of the mirror and the complicated telescope optics give rise to a nontrivial point spread function.      
      <font color=red>use the sample of a moderately-bright star to characterize PSF <font color=black>and use a polynomial to describe the viaration across the field (which is actually not accurate enough)   
    - <font color=red>PSF correction  
      <font color=black>a very important step and energective field  
      numerous methods: KSB (describe PSF as the convolution of a compact anisotropic kernel and a large isotropic kernel) / fitting of different models
    - <font color=red>Measurements of shear correlations and cosmological parameters  
      <font color=red>two point shear correlations: <font color=black>calculated simply from the galaxy positions, ellipticities, and weights(S/N)  
      <font color=red>photometric redshifts: <font color=black>based on the measured galaxy colors and other properties


- <font color=green>**Diagnostics of the lensing signal**  
<font color=black>How can we be sure that the shear is cosmological in nature instead of observational distortions?
    - <font color=red>Test the above pipeline on simulated data
      <font color=black>BUT these tests can not present the systematic effects in real data  
      The shear testing programme: STEP  
    - The corrected galaxy shapes should not correlate with the (uncorrected) shapes of stars  
      this correlation is sensitive to the model of PSF  
    - The presence of a significant <font color=red>B mode ellipticity <font color=black>indicates the existence of systematic effects  
      E mode - caused by the gravitational potential field (curl-free)
    - <font color=red>The viaration with redshift

### <font color=green>**Cosmic shear and dark energy**  
- <font color=green>**Two-point shear correlations and tomography(体层/切片摄影？)**  
  <font color=black>To quantify the lensing signal, we measure the shear correlation functions from <font color=red>galaxy shape catalogs  
  <font color=red>two point shear correlation function: <img src=".\fig\shearcorrelation.png" width="20%">, <font color=black>where $\theta=|\theta_{1}-\theta_{2}|$  
  note that i and j mark **different redshift bins** (that's why cosmic shear differs from lensing of galaxies)  
  <font color=red>+/× decomposition  
  <font color=black>+component is parallel or perpendicular to the line connecting two points  
  ×component is along 45°  
  allow the definition of rotationally invariant two-point correlations of the shear field  
  <font color=red>E/B mode decomposition  
  <font color=black>given by linear superpositions of $\xi_{+}(\theta)$ and $\xi_{×}(\theta)$  
  a more direct way is through the mass aperture variance, $M^{2}_{ap}(\theta)$ (质量孔径方差)  
  <font color=red>The shear power spectrum at angular wave number $l$  
  <font color=black>the fourier transformation of the shear correlation function  
  equivalent to the convergence power spectrum

  senstive to geometric factors and the growth of the structure → sensitive to dark energy and its possible evolution  
  powerful in testing these theories

- <font color=green>**Cross correlations and higher-order statistics**

- <font color=green>**Cosmological parameters**  
  <font color=black>Able to constrain: dark energy density $\Omega_{de}$ / equation of state paras $w_{0}$ / parameters affecting the primordial power spectrum or the growth of the structure / neutrino masses and running spectrum index / dark energy paras affecting the amplitude at different redshift (lensing tomography)

#### **<font color=green> PSF modeling**  
effective PSF includes: atmosphere PSF, optical PSF, pixel response, and charge diffusion  
  <img src="./fig/PSF.png" width="40%">  
methods removing the impact of PSF all come with the assumption that PSF is known → modeling PSF is necessary  
PSF's impact on ensemble weak lensing shear observables depends both on mean PSF size or shape errors, and their spatial correlations and the distribution of galaxy properties. 
the components of modeling PSF: use bright stars to model; interpolating to other position  
- optical PSF: vary slowly over field-of-view and exhibit a limited set of predictale patterns  
- atmosphere PSF: stochastic behavior  
- **empirical modeling**  
  choose a set of basic functions to desribe the bright star images and some to interpolate between images  
  e.g **PSFEx**  
- **principal component analysis(PCA)** 主成分分析  
  has the potential to adress both PSF modeling and interpolation  
  consider all of the survey data and identify the most important patterns  
  can determine PSF model variation as a function of focal 
plane position at higher spatial frequenc  
- **physics modeling approach**  
  ray-tracing through a physical model of the telescope optics using **Tiny Tim** http://www.stsci.edu/software/tinytim/  
  simulate and match the stars to identify the best model  
  the interpolation then use the finely-spaced grid of models  
- **combine optical modeling and atmoheric modeling**  
  can potentially include chip discontinuities  
- **higher order moments of PSF model**  
- **chromatic PSF**  
  depend on wavelenth → the effective PSF for any given object must depend on its spectral energy distribution (SED）    
  galaxy colour gradients make a single well-defined PSF for the galaxy impossible  
- **error calculating**  
  null tests  
  involve sets of stars that are not used for modeling    
  A comparison of their sizes and shapes based on second moments with the PSF model sizes and shapes at the positions of those star → make a historgram of the differences → the spacial correlation function of these errors determines how weak lensing observalbles will be biased due to PSF modeling errors.  
  PSF shape errors: 5 correlation functions
  another approach is to compare the PSF shape and size errors with the distribution of non-PSF stars
- **survey strategy**  
  substantial dithering compared to the size of the focal plane → photos of galaxies at different focal plane positions → average down the systematics due to PSF modeling errors

#### **<font color=green> Detector systematics**  
<font color=black>the problematic systematics caused by the detector non-idealities  
- **flux-dependent effeects that predominantly affect bright objects**  
  e.g non-linearity or the bright-fatter effect(brighter objects appear broader than fainter ones → a dynamic change in pixel boundaries)      
  PSF are modeled from bright stars while the measurements are dominated by faint galaxies  
  proposed methods: flat fields, detailed simulation and measurement methods  
- **defects related to galaxies orientation or position of the detector**  
  induce coherent shear signals or photometry errors  
  cause selection biases due to coherent masking patterns  
  e.g in HST, CTI(charge transfer inefficiency) causes radiation damage and imparts a preferential direction  
- It's important to understand the spacial- and time- dependence of detector effects

#### **<font color=green>Selection bias**  
<font color=black>Selection bias arises when the select or weight of galaxies depend on the galaxy shape, which is implicit  
the probability of a galaxy entering the sample (or its assigned weight) depends on its alignment with respect to the shear or PSF anisotropy direction.  
- **kinds of biases**  
  depends on shear: multiplicative biases  
  depends on the PSF shape: addictive biases  
  continuous quantities used for selection: biases only present near the boundary of the sample  
  quantities used to construct weighted average are also continuous variables: biases may be present throughout the sample  
  avoiding elongated structures like bad columns: regional biases  
- **approaches to avoid selection biases**  
    - simply avoid it  
      define detection significances and apparent sizes compared to the PSF using round kernels, so that the results are insensitive to galaxy shape    
    - estimate the magnitude through an analytic formalism and remove it  
      estimated through simulation, must take how selection biases vary based on the full range of observing conditions into consideration  
    - self-calibration approaches to shear estimation such as metacalibration  
Selection bias plays an important role in defining usefull null tests: subsets of the galaxy sample should be used to avoid systematics
